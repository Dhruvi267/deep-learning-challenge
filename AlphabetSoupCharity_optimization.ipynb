## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import pandas and read the charity_data.csv from the provided cloud URL.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 500
app_count = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = app_count[app_count < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_list = application_df['CLASSIFICATION'].value_counts()
classification_list
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_list[classification_list>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1000
class_count = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = class_count[class_count < cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, dtype=int)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df.drop('IS_SUCCESSFUL', axis =1).values
y = application_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=128, activation='relu', input_dim=X_train.shape[1]))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=64, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
history = nn.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_test, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity.h5')
```

## Optimize the Model

```python
# Create the optimized model with more neurons
nn_optimized = tf.keras.models.Sequential()

# First hidden layer with more neurons
nn_optimized.add(tf.keras.layers.Dense(units=256, activation='relu', input_dim=X_train.shape[1]))
nn_optimized.add(tf.keras.layers.Dropout(0.2))  # Dropout to avoid overfitting

# Second hidden layer with more neurons
nn_optimized.add(tf.keras.layers.Dense(units=128, activation='relu'))
nn_optimized.add(tf.keras.layers.Dropout(0.2))  # Dropout to avoid overfitting

# Third hidden layer (to make it deeper)
nn_optimized.add(tf.keras.layers.Dense(units=64, activation='relu'))

# Output layer
nn_optimized.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))  # Sigmoid for binary classification

# Check the summary of the model to confirm structure
nn_optimized.summary()
```

```python
# Define the model using the 'tanh' activation function
nn_optimized = tf.keras.models.Sequential()

# First hidden layer with tanh activation
nn_optimized.add(tf.keras.layers.Dense(units=256, input_dim=X_train.shape[1]))
nn_optimized.add(tf.keras.layers.Activation('tanh'))  # 'tanh' activation instead of ReLU
nn_optimized.add(tf.keras.layers.Dropout(0.2))  # Dropout to avoid overfitting

# Second hidden layer with tanh activation
nn_optimized.add(tf.keras.layers.Dense(units=128))
nn_optimized.add(tf.keras.layers.Activation('tanh'))  # 'tanh' activation
nn_optimized.add(tf.keras.layers.Dropout(0.2))  # Dropout to avoid overfitting

# Third hidden layer
nn_optimized.add(tf.keras.layers.Dense(units=64))
nn_optimized.add(tf.keras.layers.Activation('tanh'))  # 'tanh' activation

# Output layer
nn_optimized.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))  # Sigmoid for binary classification

# Check the model summary
nn_optimized.summary()
```

* Reduce the Epochs to 100:

```python
# Compile the optimized model (same as before)
nn_optimized.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

# Train the model with reduced epochs (100 epochs)
history_optimized_100 = nn_optimized.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

# Check the training and testing accuracy
print(f"Optimized Training Accuracy (100 epochs): {history_optimized_100.history['accuracy'][-1]}")
print(f"Optimized Testing Accuracy (100 epochs): {history_optimized_100.history['val_accuracy'][-1]}")
```

* Increase the Epochs to 300:

```python
# Compile the optimized model (same as before)
nn_optimized.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

# Train the model with increased epochs (300 epochs)
history_optimized_300 = nn_optimized.fit(X_train, y_train, epochs=300, batch_size=32, validation_data=(X_test, y_test))

# Check the training and testing accuracy
print(f"Optimized Training Accuracy (300 epochs): {history_optimized_300.history['accuracy'][-1]}")
print(f"Optimized Testing Accuracy (300 epochs): {history_optimized_300.history['val_accuracy'][-1]}")
```

```python
# Evaluate the optimized model using the test data
model_loss, model_accuracy = nn_optimized.evaluate(X_test, y_test, verbose=2)
print(f"Optimized Model Loss: {model_loss}, Optimized Model Accuracy: {model_accuracy}")
```
