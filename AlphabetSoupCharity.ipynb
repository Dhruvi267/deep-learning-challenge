## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import pandas and read the charity_data.csv from the provided cloud URL.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 500
app_count = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = app_count[app_count < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_list = application_df['CLASSIFICATION'].value_counts()
classification_list
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_list[classification_list>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = 1000
class_count = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = class_count[class_count < cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df, dtype=int)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df.drop('IS_SUCCESSFUL', axis =1).values
y = application_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=128, activation='relu', input_dim=X_train.shape[1]))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=64, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
history = nn.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_test, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('AlphabetSoupCharity.h5')
```

#### Overview:
The goal of this analysis is to create a model that can predict whether organizations funded by Alphabet Soup will be successful, based on various features like application type, income, and funding request.

#### Results:
- **Target Variable**: The target is `IS_SUCCESSFUL`, which tells us if the organization was successful with the funding.
- **Features**: We used columns like application type, affiliation, and funding amount, while excluding `IS_SUCCESSFUL`, `EIN`, and `NAME`.
- **Removed Variables**: `EIN` and `NAME` were removed since they are just identifiers and don’t help with predictions.

#### Model:
- **Neurons and Layers**: 
  - The input layer has 128 neurons with **ReLU** activation to process the features.
  - A second hidden layer with 64 neurons and **ReLU** activation to learn more complex patterns.
  - The output layer has 1 neuron with **Sigmoid** activation to predict the success rate (binary outcome).
- **Optimizer and Loss**: The model was compiled using the **Adam** optimizer and **binary cross-entropy** as the loss function, with accuracy as the evaluation metric.

#### Performance:
- The initial model achieved an accuracy of **72%**, which is below the target of **75%**.
- To improve performance, we added **dropout layers** to avoid overfitting, increased the number of neurons, and added a third hidden layer.

#### Recommendation:
- **Random Forest** could be a good alternative for this task, as it handles categorical data well and is less prone to overfitting compared to deep learning models.
